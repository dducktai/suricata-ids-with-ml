### Preprocessing

```python
# For improving performance and reduce memory-related errors
old_memory_usage = data.memory_usage().sum() / 1024 ** 2
print(f'Initial memory usage: {old_memory_usage:.2f} MB')
for col in data.columns:
    col_type = data[col].dtype
    if col_type != object:
        c_min = data[col].min()
        c_max = data[col].max()
        # Downcasting float64 to float32
        if str(col_type).find('float') >= 0 and c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
            data[col] = data[col].astype(np.float32)

        # Downcasting int64 to int32
        elif str(col_type).find('int') >= 0 and c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
            data[col] = data[col].astype(np.int32)

new_memory_usage = data.memory_usage().sum() / 1024 ** 2
print(f"Final memory usage: {new_memory_usage:.2f} MB")

# Calculating percentage reduction in memory usage
print(f'Reduced memory usage: {1 - (new_memory_usage / old_memory_usage):.2%}')

# Drop the specific column 'bwd_iat_tot'
if 'bwd_iat_tot' in data.columns:
    data.drop('bwd_iat_tot', axis=1, inplace=True)
    print("Dropped 'bwd_iat_tot' column.")

# Dropping columns with only one unique value
num_unique = data.nunique()
one_variable = num_unique[num_unique == 1]
not_one_variable = num_unique[num_unique > 1].index

dropped_cols = one_variable.index
data = data[not_one_variable]

print('Dropped columns:')
dropped_cols
```

#### Random Forest using Cross Validation

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np
import joblib
import os

# Attack Number mapping:
# 0: BENIGN
# 1: Bot
# 2: Brute Force
# 3: DDoS
# 4: DoS
# 5: Heartbleed
# 6: Infiltration
# 7: Port Scan
# 8: Web Attack

# Kiểm tra phân phối lớp ban đầu (dựa trên Attack Number)
print("\nPhân phối lớp ban đầu (Attack Number):")
print(data['Attack Number'].value_counts().sort_index().rename(index={
    0: 'BENIGN', 1: 'Bot', 2: 'Brute Force', 3: 'DDoS', 4: 'DoS',
    5: 'Heartbleed', 6: 'Infiltration', 7: 'Port Scan', 8: 'Web Attack'
}))

# Chọn đặc trưng và nhãn
X = data[pos_corr_features]  # Sử dụng pos_corr_features
y = data['Attack Number']    # Sử dụng Attack Number làm nhãn

# Khởi tạo StratifiedKFold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# Danh sách lưu kết quả
accuracies = []
precisions = []
recalls = []
f1_scores = []
conf_matrices = []

# Cross-validation
print("\nBắt đầu cross-validation...")
for fold, (train_idx, test_idx) in enumerate(kfold.split(X, y), 1):
    print(f"\nFold {fold}:")

    # Chia dữ liệu
    X_train = X.iloc[train_idx]
    y_train = y.iloc[train_idx]
    X_test = X.iloc[test_idx]
    y_test = y.iloc[test_idx]

    # Áp dụng SMOTE trên tập train
    smote = SMOTE(sampling_strategy={
      5: 500,   # Infiltration
      6: 500,   # Heartbleed
    }, random_state=42, k_neighbors=3)
    X_train_upsampled, y_train_upsampled = smote.fit_resample(X_train, y_train)

    # Áp dụng undersampling cho BENIGN trên tập train
    under = RandomUnderSampler(sampling_strategy={0: 200000}, random_state=42)  # BENIGN
    X_train_balanced, y_train_balanced = under.fit_resample(X_train_upsampled, y_train_upsampled)

    # Huấn luyện mô hình
    rf = RandomForestClassifier(n_estimators=100, max_depth=10, max_features='sqrt', random_state=0)
    rf.fit(X_train_balanced, y_train_balanced)

    # Dự đoán trên tập test
    y_pred = rf.predict(X_test)

    # Tính toán metrics
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=[
        'BENIGN', 'Bot', 'Brute Force', 'DDoS', 'DoS', 'Heartbleed', 'Infiltration', 'Port Scan', 'Web Attack'
    ], output_dict=True)

    # Lưu metrics
    accuracies.append(accuracy)
    precisions.append(report['weighted avg']['precision'])
    recalls.append(report['weighted avg']['recall'])
    f1_scores.append(report['weighted avg']['f1-score'])
    conf_matrices.append(confusion_matrix(y_test, y_pred))

    # In kết quả fold
    print(f"Độ chính xác: {accuracy:.4f}")
    print("\nBáo cáo phân loại:")
    print(classification_report(y_test, y_pred, target_names=[
        'BENIGN', 'Bot', 'Brute Force', 'DDoS', 'DoS', 'Heartbleed', 'Infiltration', 'Port Scan', 'Web Attack'
    ]))

# Báo cáo tổng kết
print("\nTổng kết Cross-Validation (5 folds):")
print(f"Độ chính xác trung bình: {np.mean(accuracies):.4f} (±{np.std(accuracies):.4f})")
print(f"Precision trung bình: {np.mean(precisions):.4f} (±{np.std(precisions):.4f})")
print(f"Recall trung bình: {np.mean(recalls):.4f} (±{np.std(recalls):.4f})")
print(f"F1-score trung bình: {np.mean(f1_scores):.4f} (±{np.std(f1_scores):.4f})")

# Tính ma trận nhầm lẫn trung bình
mean_conf_matrix = np.mean(conf_matrices, axis=0).astype(int)
print("\nMa trận nhầm lẫn trung bình:")
print(mean_conf_matrix)

# Huấn luyện mô hình cuối cùng trên toàn bộ dữ liệu sau cân bằng
print("\nHuấn luyện mô hình cuối cùng trên toàn bộ dữ liệu...")
X_upsampled, y_upsampled = smote.fit_resample(X, y)
X_balanced, y_balanced = under.fit_resample(X_upsampled, y_upsampled)
rf_final = RandomForestClassifier(n_estimators=100, max_depth=10, max_features='sqrt', random_state=0)
rf_final.fit(X_balanced, y_balanced)

# Tạo folder lưu model nếu chưa có
save_dir = '/content/drive/MyDrive/idps-with-ml/model'
os.makedirs(save_dir, exist_ok=True)

# Lưu mô hình
model_path = os.path.join(save_dir, 'random_forest.pkl')
joblib.dump(rf_final, model_path)
print(f'\n✅ Mô hình đã lưu tại: {model_path}')
```
